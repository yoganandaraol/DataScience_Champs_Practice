```python
from sklearn import preprocessing
### label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
### Encode labels in column 'Sex'. 
df_insurance['sex']= label_encoder.fit_transform(df_insurance['sex'])
df_insurance['smoker']= label_encoder.fit_transform(df_insurance['smoker'])
df_insurance['region']= label_encoder.fit_transform(df_insurance['region'])                                                   
df_insurance.head()
```

```python
#Function for returning BMI category based on BMI value
#r --> Each row all the series in df
def getbmicategory(r):
    if (r.bmi > 30) & (r.charges > 30000.0):
        return 'High'
    else:
        return 'Low'
```

```python
df_insurance['bmi_cat'] = df_insurance.apply(getbmicategory, axis=1)
```

```python
df_insurance.head()
```

```python
sns.scatterplot(data=df_insurance, x="bmi", y="charges", hue="bmi_cat")
```

```python
#Function for returning BMI Intense category based on BMI value and Smoking Habbit
#r --> Each row all the series in df
def getBmiIntenseCategory(r):
    if (r.bmi > 30) & (r.smoker == 1):
        return 'High'
    else:
        return 'Low'
```

```python
#Creating a new Column as bmi_Intense_cat using Function getBmiIntenseCategory
#This new column will not have dependency on Target variable (Charges)
df_insurance['bmi_Intense_cat'] = df_insurance.apply(getBmiIntenseCategory, axis=1)
```

```python
df_insurance.head()
```

```python
df_insurance.tail()
```

```python
sns.scatterplot(data=df_insurance, x="bmi", y="charges", hue="bmi_Intense_cat")
```

* __People with BMI>30 and also with Smoking Habbit are Charged Premium Charges >30000__

```python
df_insurance.drop('bmi_cat', axis=1, inplace=True)
```

```python
df_insurance.head()
```

```python
df_insurance['bmi_Intense_cat']= label_encoder.fit_transform(df_insurance['bmi_Intense_cat'], ) 
df_insurance.head()
```

```python
sns.heatmap(df_insurance.corr(), annot= True)
```

```python
#Inversing bmi_Intense_cat: Low=0 & High=1
def inverseBmiIntense(r):
    if (r.bmi_Intense_cat == 1):
        return 0
    else:
        return 1
```

```python
df_insurance['bmi_Intense_cat'] = df_insurance.apply(inverseBmiIntense, axis=1)
df_insurance.head()
```

```python
sns.heatmap(df_insurance.corr(), annot= True)
```

* __Observed Region(Corr. value close to Zero) is not influencing the charges__

```python
# Import Scikit learn packages 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
depCols = set(df_insurance.columns) - set('charges')
```

```python
# depCols
```

```python
dep_cols = ['age', 'sex', 'bmi', 'children', 'smoker', 'region','bmi_Intense_cat']
tar_cols = ['charges']
```

```python
X = df_insurance[dep_cols]
y = df_insurance[tar_cols]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
lr = LinearRegression()  # instantiate Linear Regression model
```

```python
model = lr.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Check the error

```python
from sklearn import metrics
from sklearn.metrics import r2_score # For 
```

```python
# MAE
MAE = metrics.mean_absolute_error(y_test['charges'], y_pred_df['charges'])
```

```python
MSE = metrics.mean_squared_error(y_test['charges'], y_pred_df['charges'])
```

```python
RMSE = np.sqrt(MSE)
```

```python
R_Sq = r2_score(y_test['charges'], y_pred)
```

```python
Adj_RSq = 1- (1-R_Sq)*(len(y_test)-1)/(len(y_test) - X_train.shape[1] -1)
Adj_RSq
```




Drawback - When the correlation among the fetures are weak and features are more the __R-Sq__ value will be still increasing.
 __Adj R-Sq__ overcomes this issue by adding penality.

* Penality Factor


```python
Adj_RSq2 = 1- (1-R_Sq)*(len(y_test)-1)/(len(y_test) - len(X_train.columns) -1)
Adj_RSq2
```

```python
print('MAE - ', MAE)
print('MSE - ', MSE)
print('RMSE - ', RMSE)
print('R-Sq - ', R_Sq)
print('Adj R-Sq - ', Adj_RSq)
```
